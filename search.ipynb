```python
import skvideo.io
import skvideo.datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
videodata_bigbuckbunny = skvideo.io.vread(skvideo.datasets.bigbuckbunny())
```

```python
videodata_bikes = skvideo.io.vread(skvideo.datasets.bikes())
```

```python
from indexer import VideoIndexer
hist_index_bigbuckbunny, phash_index_bigbuckbunny = VideoIndexer("bigbuckbunny.mp4", videodata_bigbuckbunny).index()
```

```python
hist_index_bikes, phash_index_bikes = VideoIndexer("bikes.mp4", videodata_bikes).index()
```

```python
fname2videodata = {
    "bigbuckbunny.mp4": videodata_bigbuckbunny,
    "bikes.mp4": videodata_bikes
}
```

```python
hist_index = hist_index_bigbuckbunny + hist_index_bikes
phash_index = phash_index_bigbuckbunny + phash_index_bikes
```

```python
import pickle
print(len(pickle.dumps((phash_index, hist_index))))
print(len(phash_index), len(videodata_bigbuckbunny) + len(videodata_bikes))
print(len(hist_index), len(videodata_bigbuckbunny) + len(videodata_bikes))
```
